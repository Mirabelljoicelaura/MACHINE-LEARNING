# PRAKTIKUM 1 - Pengolahan Data - Data Terstruktur

**Langkah 1 - Load Data**
Import library yang dibutuhkan kemudian, load data kedalam DataFrame Pandas.

```python
import pandas as pd
data = 'Titanic-Dataset.csv'
df = pd.read_csv(data)
```

Tampilkan data teratas.

```python
df.head()
```

**Langkah 2 - Pengecekkan Data** <br>
Lakukan pengecekan informasi mengenai data

```python
df.info()
```

Kode program dibawah digunakna untuk mengetahui jumlah data yang hilang untuk setiap kolom.

```python
df.isnull().sum()
```

**Langkah 3 - Imputasi**
Pada langkah ini kita akan melakukan imputasi terdapat data yang hilang pada "Age", "Cabin", dan "Embarked".

```python
# Age : adalah data nominal, maka menggunakan strategi mean.
df['Age'].fillna(value=df['Age'].mean(), inplace=True)

# Cabin adalah informasi terkait dengan nomor sehingga kita akan mensubstitusi data yang hilang dengan informasi "DECK"
df['Cabin'].fillna(value="DECK", inplace=True)

# Embarked adalah informasi lokasi embarkasi penumpang sehingga dapat menggunakan modus untuk substitusi data yang hilang
df['Embarked'].fillna(value=df['Embarked'].mode, inplace=True)
```

**Langkah 4 - Validasi Hasil** <br>
Lakukan pengecekan kembali dengan df.info()

```python
df.info()
```

```python
df.isnull().sum()
```

# Praktikum 2 - Seleksi Fitur, Encoding, dan Standarisasi

**Langkah 0 - Load Library**

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

**Langkah 1 - Load Data**

```python
dpath = 'Titanic-Dataset-fixed.csv'
df = pd.read_csv(dpath)
df.head()
```

**Langkah 2 - Slice Data**

```python
df = df[['Survived', 'Pclass', 'Age', 'Sex', 'Cabin']]
df.head()
```

**Langkah 3 - Encoding**
<br>
Melakukan encoding pada data-data kategorikal, yaitu "Sex" dan "Cabin".Label Encoding tidak mengurutkan data terlebih dahulu.

```python
# membuat object dari label encoder
le = LabelEncoder()
#proses encoding
df['Sex'] = le.fit_transform(df['Sex'])
#proses encoding
df['Cabin'] = le.fit_transform(df['Cabin'])
```

**Langkah 4 - Verifikasi Hasil**

```python
df.head()
```

Langkah 5 - Standarisasi

```python
std = StandardScaler()
df['Age'] = std.fit_transform(df[['Age']])
```
